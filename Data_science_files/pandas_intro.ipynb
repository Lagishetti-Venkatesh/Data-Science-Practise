```python
#importing libraries
import pandas as pd 

#reading files.
df = pd.read_csv('salaries.csv')
print(df)

#print shape, column attributes   and info, describe methods
print(df.shape)
print(df.columns)
print(df.info())
print(df.describe())
print(df.describe(include = 'object'))

#printing the sorted values
print(df.sort_values(by = ['Salary'], axis = 0, ascending= True ))

#printing largest and smallest
print(df.nlargest(10, 'Salary'))
print(df.nsmallest(10, 'Salary'))
```
