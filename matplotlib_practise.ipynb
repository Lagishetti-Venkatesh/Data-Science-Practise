```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi , 256, endpoint = True)

S, C = np.sin(x),np.cos(x)

fig = plt.figure(figsize = (10,5), dpi = 80)
plt.plot(x, S,color = "red" , linewidth=3, linestyle = '-', label= 'Sine_Wave' )
plt.plot(x, C, color='green', linewidth= 2, linestyle= '--', label= "Cosine_wave")
plt.xlim(-4.0, 4.0)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.ylim(-1, 1)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

plt.legend(loc='upper left', frameon = True)
plt.title("Graph of Wave movement with Sine and Cosine functions")
plt.xlabel("Time , t")
plt.ylabel("Position , P")

plt.grid(color='gray', linestyle='-', linewidth =0.2)
ax = plt.gca()

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.spines['top'].set_color('None')
ax.spines['right'].set_color('None')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
#fig.savefig('my_figure_2.png')


plt.show()


```

```python
#scatter plots
N = 100
 
# Generate 2 different arrays
x = np.random.rand(N)
y = np.random.rand(N)
 
fig = plt.figure(figsize=(9, 6))
 
# Assign random colors and variable sizes to the bubbles
colors = np.random.rand(N)
print((colors*20*np.pi))
print((colors*20*np.pi)**2)
print('\n\n')
area = np.pi * (20 * np.random.rand(N))**2 # 0 to 20 point radii
# Scatter plot on x-y coordinate with the assigned size and color
plt.scatter(x, y, s=area, c=colors, alpha=0.7)
 
plt.show()
```
