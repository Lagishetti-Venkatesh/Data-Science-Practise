```python
#finding the skew distributions
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv('forestfires.csv')
v = df.columns

plt.hist(df[v[4]], bins = 10)
plt.title('Left-tailed ditribtuions')
plt.show()


plt.hist(df[v[5]], bins = 10)
plt.title('Right-tailed Distribution ')
plt.show()
```

```python
#converting the skew into normal distribution
df[v[4]] =np.log(df[v[4]]) 
df[v[5]] =np.log(df[v[5]])
plt.hist(df[v[4]], bins = 10)
plt.show()
plt.hist(df[v[5]], bins = 10)
plt.show()
```

```python
from sklearn import preprocessing

df_temp_1 = df[[v[4]]]
df_temp_2 = df[[v[5]]]
# Normalization of numeric variables - not to be implemented on df, so using df_temp
df_temp_1 = preprocessing.normalize(df_temp_1)
df_temp_2 = preprocessing.normalize(df_temp_2)
print(df_temp_1, end='\n\n')
print(df_temp_2, end='\n\n')
```
