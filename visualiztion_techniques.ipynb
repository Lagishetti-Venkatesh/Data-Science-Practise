```python
# multiple histograms 

#importing libraries
import numpy as np 
import matplotlib.pyplot as plt


np.random.seed(2)
#creating data for multiple histograms
usa = 30 * np.random.randn(1000)+100
uk = 20 * np.random.randn(1000)+100

#plotting Datasets
plt.hist([usa, uk], label = ['US', 'UK'], histtype = 'stepfilled', alpha = 0.5, bins = 30, ec = 'black')

plt.legend(loc = 'best')
plt.title("The Superimposing of Datasets")
plt.xlabel('Salary Interval')
plt.ylabel('Frequency of Salary')

plt.show()

# USA is platykurtic that is light tailed and wider distribution of points which says that USA have higher amount of salaries than UK
#Uk is leptoKurtic that is heavilty tailed and more points are concentrated at center which means most of them have salaries in between 100$ to 125$.
```

```python
#Barstacked 
#plotting Datasets
plt.hist([usa,uk], label=['USA','UK'], histtype='barstacked', bins=30, ec='black')
plt.title("Histogram – Height of Incomes")    
plt.ylabel("Frequency") 
plt.xlabel("Income Bins")
plt.legend(loc='upper right')
plt.show()
# USA is platykurtic that is light tailed and wider distribution of points which says that USA have higher amount of salaries than UK
#Uk is leptoKurtic that is heavilty tailed and more points are concentrated at center which means most of them have salaries in between 100$ to 125$.
```

```python
# Side by side histogram
plt.hist([usa, uk], label=['USA','UK'], histtype='bar', bins=20, ec='black')
plt.title("Histogram – Height of Incomes")    
plt.ylabel("Frequency") 
plt.xlabel("Income Bins")
plt.legend(loc='upper right')
plt.show()
```

```python
from sklearn.datasets import load_wine
import pandas as pd
 
pd.set_option('display.width', 120)
pd.set_option('display.max_columns', 14)
 
wine = load_wine()
data = pd.DataFrame(data=wine.data, columns=wine.feature_names)
data = data[['alcohol', 'malic_acid', 'ash', 'magnesium', 'proanthocyanins', 'proline', ]]
 
```

```python
ash = data.ash
pro = data.proanthocyanins

plt.hist([ash, pro], label = ['Ash', 'Proanthocyanins'], histtype = 'stepfilled',bins = 30, alpha = 0.5, ec = 'black')
plt.legend(loc = 'upper right')
plt.title('Contents of Ash and proanthocyanins')
plt.xlabel('bins of content')
plt.ylabel('Frequency or Content')
plt.show()
```

```python

from sklearn.datasets import load_boston
import pandas as pd
 
# load dataset and convert into pandas dataframe
boston = load_boston()
data = pd.DataFrame(data=boston.data, columns=boston.feature_names)

dis = data.DIS

import numpy as np
import pandas as pd
outliers=[]
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
outlier_datapoints = detect_outlier(dis)
print(outlier_datapoints)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns 

# generate data 
np.random.seed(2)
 
# Group 1 has mean 165 and stddev 10
df1 = pd.DataFrame()
df1['Height'] = np.round((10 * np.random.randn(500) + 165), 0)
df1['Group'] = 'Group1'
 
# Group 2 has mean 170 and stddev 12
df2 = pd.DataFrame()
df2['Height'] = np.round((12 * np.random.randn(500) + 170), 0)
df2['Group'] = 'Group2'
df = df1.append(df2)
sns.boxplot(data = df, x= 'Height', y = 'Group')
```

```python
# randomly assign Basketball_Player status
df['Basketball_Player'] = np.random.randint(0, 2, df.shape[0])
# use the 'hue' parameter to add colors based on Basketball_Player status
sns.boxplot(x='Height', y='Group', hue='Basketball_Player', data=df)
plt.show()
```
